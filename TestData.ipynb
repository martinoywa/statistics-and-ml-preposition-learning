```python
import pickle
import itertools
import numpy as np
```

```python
with open('SentencesTest.txt', 'r') as f:
    test_sentences = f.readlines()
```

```python
test_sentences_clean = []

for sentence in test_sentences:
    temp_sentence = []
    for word in sentence.split():
        temp_sentence.append(word.split('/')[0])
    test_sentences_clean.append(temp_sentence)
```

```python
test_sentences_clean
```

```python
sentence_list = list(set(itertools.chain.from_iterable(test_sentences_clean)))
```

```python
sentence_list[:10]
```

```python
with open('200/model.txt', 'r') as f:
    embeddings = f.readlines()
```

```python
del embeddings[0]
```

```python
word_embeddings_dict = dict()
```

```python
for embedding in embeddings:
    for word in sentence_list:
        if embedding.startswith(word) and word not in word_embeddings_dict.keys():
            word_embeddings_dict[word] = list(map(float, embedding.split()[1:]))
        else:
            pass
```

```python
sentences_test_embeddings = []
```

```python
for sentence in test_sentences:
    sum_temp = np.array([0.0]*300)
    for word in sentence:
        if word in word_embeddings_dict.keys():
            sum_temp += np.array(word_embeddings_dict[word])
        else: pass
    sentences_test_embeddings.append(list(sum_temp))
```

```python
sentences_test_embeddings
```

```python
# save the embeddings list
with open('Test-Vector-Embeddings.pk', 'wb') as f:
    pickle.dump(sentences_test_embeddings, f)
```
